```python
# some functions
def replace(ls):
    return [i.replace(' ','_')for i in ls]
def method_category(wiki):
    return [j['title'] for i,j in enumerate(wiki['query']['categorymembers']) if 'Category:' in j['title']]
def retrieve(method_list):
    baseurl = "http://en.wikipedia.org/w/api.php/?"
    action = "action=query"
    title='cmtitle='+method_list+'&cmlimit=500' 
    content = "list=categorymembers"
    rvprop ="rvprop=content"
    dataformat = "format=json"
    query = "{}{}&{}&{}&{}".format(baseurl, action, dataformat, content,  title)
    response=urllib.request.urlopen(query)
    wikisource = response.read()
    wiki= json.loads(wikisource)
    return wiki
```
